```python
# step 1:Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
#step 2:loading dataset
url ="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(url)
df.head(10)
```

```python
# step 3: Visualize Data
plt.scatter(df['Hours'],df['Scores'])
plt.title("Hours vs Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid()
plt.show()
```

```python
#step 4: prepare data
x=df[['Hours']]
y=df['Scores']
```

```python
# step 5: Train/test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)
```

```python
#step 6 : Train Linear Regression Model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#step 7:predict
y_predict = model.predict(x_test)
```

```python
#step 8: Evaluate Model
print("R^2 score:",r2_score(y_test,y_predict))
print("Mean absoulte Error:",mean_absolute_error(y_test,y_predict))
```

```python
#step 9:Visualize Regression line
plt.scatter(x,y,color='blue')
plt.plot(x,model.predict(x),color='red') #REGRESSION LINE
plt.title("Regression Line - Student scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
#step 10: Predict for custom inp
hours=[[9.25]]
predicted_score=model.predict(hours)
print(f"Predicted score for studying{hours[0][0]} hours={predicted_score[0]:.2f}")
```
