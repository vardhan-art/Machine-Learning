# SIMPLE LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample dataset:Temp vs Ice Cream sales
np.random.seed(0)
temp=np.random.randint(20,40,50)
sales= 2.5*temp + np.random.normal(0,5,50)
df=pd.DataFrame({'Temperature': temp ,'Ice Cream sales':sales})
df.head(10)
```

```python
#MODEL
X=df[['Temperature']]
Y=df['Ice Cream sales']
model=LinearRegression()
model.fit(X,Y)
```

```python
#Plot
plt.scatter(X,Y,color="green")
plt.plot(X,model.predict(X),color="red")
plt.xlabel("Temperature(C)")
plt.ylabel("Ice Cream sales")
plt.title("Simple Linear Regression")
plt.show()
```

2)MULTIPLE LINEAR REGRESSION PRACTICAL

```python
np.random.seed(1)
study_hour=np.random.randint(1,10,50)#(mean,std,no. of values)
attendance=np.random.randint(50,100,50)
sleep_hour=np.random.randint(5,9,50)
score=(5 *study_hour + 0.5 * attendance + 2 * sleep_hour+np.random.normal(0,10,50))
```

```python
df=pd.DataFrame({'study_hour': study_hour,'attendance':attendance,'sleep_hour':sleep_hour,'score':score})
df.head(10)
```

```python
X=df[['study_hour','attendance','sleep_hour']]
Y=df['score']
model=LinearRegression()
model.fit(X,Y)
```

```python
print("Coefficients",model.coef_)
print("Intercept :",model.intercept_)
```

```python
#Homework
plt.scatter(study_hour,score,color="green")
plt.plot(study_hour,model.predict(study_hour),color="red")
plt.xlabel("study_hour")
plt.ylabel("score")
plt.title("Multiple  Linear Regression")
plt.show()
```
