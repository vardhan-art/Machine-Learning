# POLYNOMIAL REGRESSION

```python
#step 1 : import Librarie
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#step 2 : Load Dataset
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(url)
df.head()
```

```python
#step 3: Prepare the dta
x=df[['Hours']].values #Feature
y=df['Scores'].values  #Target
```

```python
#step 4: Polynomial Feature Transformation(Degree =2)
ploy = PolynomialFeatures(degree=2)
x_poly = ploy.fit_transform(x)
```

```python
#step 5
model=LinearRegression()
model.fit(x_poly,y)
```

```python
#step 6: Evaluate Model
y_pred=model.predict(x_poly)
print("R^2 score:",r2_score(y,y_pred))
print("Mean absoulte Error:",mean_absolute_error(y,y_pred))
```

```python
#step 7: Plotting the result
plt.scatter(x,y,color='blue',label="Actual data")
#Sorting for smooth curve
x_sorted = np.sort(x,axis=0)
y_sorted_pred=model.predict(ploy.transform(x_sorted))
plt.plot(x_sorted,y_sorted_pred,color='red',label='Polynomial regression')
plt.title("Polynomial Regression (Hours vs Score)")
plt.xlabel("Hours stuied")
plt.ylabel("Score")
plt.legend()
plt.show()
```
